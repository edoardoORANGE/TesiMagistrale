```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Spectrum = pd.read_csv('Validazione_Lu177/Spectrum_vals.csv',sep=',')
Ion = pd.read_csv('Validazione_Lu177/Ion_vals.csv',sep=',')
```

```python
Ch = 1.271319178358  
NCh = 1.55438878911315
Bminus = 1.000000049


TOT=NCh+Ch+Bminus

print(TOT)
```

```python
plt.plot(Spectrum['Distance_(pixels)'], Spectrum['Gray_Value']*TOT, label='Spectrum')
plt.plot(Ion['Distance_(pixels)'], Ion['Gray_Value'], label='Ion')

plt.title('Profilo di Dose Ion vs Spectrum - Lu177')
plt.xlabel('Distance [pixels]')
plt.ylabel('Grey Value [Gy/Bq s]')
plt.ylim(0,1.2*max(Ion['Gray_Value']))
plt.grid()
plt.legend()
plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0), useMathText=True)
#plt.savefig("DoseProfiles_Lu177.png")
plt.show()
```

```python
perc = (Ion['Gray_Value']-(Spectrum['Gray_Value']*TOT))*100/Ion['Gray_Value']

diff = (Ion['Gray_Value']-(Spectrum['Gray_Value']*TOT))

plt.plot(Ion['Distance_(pixels)'], np.abs(perc), label='Ion-Sbagliato')

plt.title('Differenza Relativa Percentuale Ion-Spectrum')
plt.xlabel('Distance [pixels]')
plt.ylabel('|(Ion-Spectrum)/Ion| [%]')
plt.legend(loc='upper right')
plt.xlim(20,50)
plt.ylim(-5,20)
plt.grid()

#plt.savefig("RelDiff_Lu177.png")
plt.show()
```

```python
plt.plot(Ion['Distance_(pixels)'], diff, label='Ion-Spectrum')



plt.title('Differenza Ion-Spectrum')
plt.xlabel('Distance [pixels]')
plt.ylabel('Ion-Spectrum [Gy/Bq s]')
plt.legend(loc='upper right')
#plt.xlim(20,50)
plt.ylim(-3e-7,3e-7)
plt.grid()

#plt.savefig("Diff_Lu177.png")
plt.show()
```
